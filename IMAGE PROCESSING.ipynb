```python
#Flattening of images : heading
#convert m x n into 1 x mn
#eg 100 x 200 = 1 x 20000

five= cv2.imread('five.jpg')
seven= cv2.imread('seven.png')
two= cv2.imread('two.jpg')
```

```python
seven.shape
```

```python
two.shape
```

```python
five.shape
```

```python
five_g= cv2.cvtColor(five,cv2.COLOR_BGR2GRAY)
seven_g= cv2.cvtColor(seven,cv2.COLOR_BGR2GRAY)
two_g= cv2.cvtColor(two,cv2.COLOR_BGR2GRAY)
```

```python
print(five_g.shape)
print(seven_g.shape)
print(two_g.shape)
```

```python
#lets plot these images
plt.imshow(five_g, cmap='gray')
```

```python
#lets plot these images
plt.imshow(seven_g, cmap='gray')
```

```python
#lets plot these images
plt.imshow(two_g, cmap='gray')
```

```python
# let's resize all three images
five_g_r = cv2.resize(five_g,(100,100))
seven_g_r = cv2.resize(seven_g,(100,100))
two_g_r = cv2.resize(two_g,(100,100))
```

```python
print(five_g_r.shape)
print(two_g_r.shape)
print(seven_g_r.shape)
```

```python
# print this resize image
plt.imshow(two_g_r, cmap='gray')
```

```python
#Flat the image now - heading

five_flat= five_g_r.reshape(100*100)
seven_flat= seven_g_r.reshape(100*100)
two_flat= two_g_r.reshape(100*100)
```

```python
print(five_flat.shape)
print(seven_flat.shape)
print(two_flat.shape)
```

```python
five_flat
```

```python
# let's stack all vectors of all 3 images on top of each other

flattened_stack = np.stack([five_flat,seven_flat,two_flat],axis=0)
```

```python
flattened_stack.shape
```

```python
# write flattened image as a df
import pandas as pd
flattened_df= pd.DataFrame(flattened_stack)
```
